```python

#calculating the total traffic time 
df['total_traffic'] = df['Total UL (Bytes)'] + df['Total DL (Bytes)']

# Grouping by MSISDN/Number and aggregating
metrics = df.groupby('MSISDN/Number').agg({
    'Nb of sec with Vol DL < 6250B': 'sum',  
    'Dur. (ms)': 'sum',                     # Total session duration
    'total_traffic': 'sum'                  # Total traffic
}).reset_index()

# Rename columns for clarity
metrics.rename(columns={
    'Nb of sec with Vol DL < 6250B': 'sessions_frequency',
    'Dur. (ms)': 'session_duration'
}, inplace=True)

print(metrics.head())

# Sort and get top 10 customers for each metric
top_sessions = metrics.nlargest(10, 'sessions_frequency')
top_duration = metrics.nlargest(10, 'session_duration')
top_traffic = metrics.nlargest(10, 'total_traffic')

print("printing the top 10....")


# Display top data as a table
print("Top 10 Sessions Frequency:")
display(top_sessions)

print("Top 10 Session Duration:")
display(top_duration)

print("Top 10 Total Traffic:")
display(top_traffic)

```

## Normalization and K-means Clustering

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
```

```python

# Normalize the data
scaler = MinMaxScaler()
normalized_data = scaler.fit_transform(metrics[['sessions_frequency', 'session_duration', 'total_traffic']])

# K-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(normalized_data)
metrics['cluster'] = clusters

# Elbow method
wcss = []
for k in range(1, 11):
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(normalized_data)
    wcss.append(km.inertia_)

plt.plot(range(1, 11), wcss, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('Elbow Method')
plt.show()
```

```python

# Assuming df is your dataset
# Calculate total traffic for each application
df['Youtube Traffic'] = df['Youtube DL (Bytes)'] + df['Youtube UL (Bytes)']
df['Netflix Traffic'] = df['Netflix DL (Bytes)'] + df['Netflix UL (Bytes)']
df['Gaming Traffic'] = df['Gaming DL (Bytes)'] + df['Gaming UL (Bytes)']
df['Other Traffic'] = df['Other DL (Bytes)'] + df['Other UL (Bytes)']

# Aggregate traffic per user (MSISDN)
agg_app_traffic = df.groupby('MSISDN/Number').agg({
    'Youtube Traffic': 'sum',
    'Netflix Traffic': 'sum',
    'Gaming Traffic': 'sum',
    'Other Traffic': 'sum'
}).reset_index()

# Print aggregated traffic
print(agg_app_traffic.head())



```

##  Top 10 Users Per Application

```python
# Top 10 users for each application
top_10_youtube = agg_app_traffic.nlargest(10, 'Youtube Traffic')
top_10_netflix = agg_app_traffic.nlargest(10, 'Netflix Traffic')
top_10_gaming = agg_app_traffic.nlargest(10, 'Gaming Traffic')
top_10_other = agg_app_traffic.nlargest(10, 'Other Traffic')

print("Top 10 YouTube Users:")
print(top_10_youtube)
```

```python


# Prepare data for visualization
top_apps = ['Youtube Traffic', 'Netflix Traffic', 'Gaming Traffic']
app_totals = agg_app_traffic[top_apps].sum().sort_values(ascending=False)

# Plot total traffic per application
plt.figure(figsize=(8, 6))
app_totals.plot(kind='bar', color=['blue', 'red', 'green'])
plt.title('Total Traffic for Top 3 Applications')
plt.xlabel('Applications')
plt.ylabel('Total Traffic (Bytes)')
plt.xticks(rotation=45)
plt.show()
```
